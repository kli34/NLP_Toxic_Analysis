# Error Analysis - Racism (FN)

```python
#import dataset Racism Log 2
race_log2 = pd.read_csv("logistic_racism2.csv")
race_log2
def CLASS(Pred_racism):
  
    if Pred_racism >= 0.5:
        return 1
    if Pred_racism < 0.5:
        return 0

        
race_log2['Pred_racism'] = race_log2['Pred_racism'].apply(CLASS)
dropvar = ['Unnamed: 0', 'id.1']
race_log2 = race_log2.drop(dropvar, axis = 1)
# race_log2
```

```python
#import dataset Racism RF 2
race_rf2 = pd.read_csv("rf_racism2.csv")
race_rf2
def CLASS(Pred_racism):
  
    if Pred_racism >= 0.5:
        return 1
    if Pred_racism < 0.5:
        return 0

        
race_rf2['Pred_racism'] = race_rf2['Pred_racism'].apply(CLASS)
dropvar = ['Unnamed: 0', 'id.1']
race_rf2 = race_rf2.drop(dropvar, axis = 1)
# race_rf2
```

```python
racism_error = pd.merge(race_log2, race_rf2, how='inner', on='id')
racism_error1 = racism_error[racism_error['Pred_racism_x'] > racism_error['Pred_racism_y']]
racism_error1 = racism_error1[['comment_text_x', 'comment_text_y', 'racism_x', 'Pred_racism_x', 'racism_y', 'Pred_racism_y']]
racism_error1.query('racism_x == 1 & Pred_racism_x == 1')
```
